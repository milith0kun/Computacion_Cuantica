# Prueba de Conexión a IBM Quantum
Este notebook carga el API Key desde `apikey.json` y prueba la conexión a los servidores de IBM Quantum para listar los backends (computadoras cuánticas y simuladores) disponibles.

```python
import json
from qiskit_ibm_runtime import QiskitRuntimeService

# 1. Leer el API key desde el archivo
with open('../apikey.json', 'r') as f:
    data = json.load(f)
    api_key = data['apikey']

print("API Key cargada exitosamente.")
```

```python
# 2. Autenticarse en IBM Quantum
try:
    service = QiskitRuntimeService(channel="ibm_quantum_platform", token=api_key)
    print("\n✅ Conectado exitosamente a IBM Quantum!")
    
    # 3. Obtener y listar los backends disponibles
    backends = service.backends()
    print(f"\nBackends disponibles ({len(backends)}):\n")
    for b in backends:
        print(f"- {b.name} (Qubits: {b.num_qubits})")
        
except Exception as e:
    print("\n❌ Error al conectar con IBM Quantum:")
    print(e)
```
